***

```python
from typing import Protocol
```

```python
class Printable(Protocol):
    pages: int
    
    def print_item(self):
        pass
    
    def recycle(self):
        pass
    
class Book:
    pages: int
        
    def __init__(self, title: str):
        self.title = title
        
    def print_item(self):
        print(f'Printing book: {self.title}')
        
    def recycle(self):
        print(f'Recycling: {self.title}')
        

class Magazine:
    pages: int
        
    def __init__(self, title: str):
        self.title = title
        
    def print_item(self):
        print(f'Printing book: {self.title}')
        
    def recycle(self):
        print(f'Recycling: {self.title}')
        
def print_printable(printable: Printable):
    printable.print_item()
    

book: Printable = Book('Python')
print_printable(book)
    
magazine: Printable = Magazine('Pythonistas')
print_printable(magazine)
```

<div id="Property_and_cached_property">
    <h2>Property and cached property</h2>
    

    
To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>
</div>

***

```python
from functools import cached_property
import time
```

```python
# start from
class ExpensiveComputation:
    
    def __init__(self, value):
        self.value = value
        
    def computation_result(self):
        print('calculating result ...')
        time.sleep(3)
        self.value *= 5
        return self.value
        
ec = ExpensiveComputation(10)
print(ec.computation_result())
print(ec.computation_result())
print(ec.computation_result())
```

```python
# as a property
class ExpensiveComputation:
    
    def __init__(self, value):
        self.value = value
     
    @property
    def computation_result(self):
        print('calculating result ...')
        s = time.time()
        time.sleep(3)
        self.value *= 5
        try:
            return self.value
        finally:
            e = time.time()
            t = e - s
            print(f'Time taken: {t}')
            
ec = ExpensiveComputation(10)
print(ec.computation_result)
print(ec.computation_result)
print(ec.computation_result)
```

### If we know once the computation is calculated it will never change, we can use a cached_property instead

- below example it only had to calulate it once why is why time is only returned once, for the 2nd and third call it just returned the previous output (cache)

```python
# as a cached_property
class ExpensiveComputation:
    
    def __init__(self, value):
        self.value = value
     
    @cached_property
    def computation_result(self):
        print('calculating result ...')
        s = time.time()
        time.sleep(3)
        try:
            return self.value * 5
        finally:
            e = time.time()
            t = e - s
            print(f'Time taken: {t}')

ec = ExpensiveComputation(10)
print(ec.computation_result)
print(ec.computation_result)
print(ec.computation_result)

# Note if you try and change the value now though it will return the wrong value as it will just return the cached value
print('*'*20)
ec.value = 100
print(ec.computation_result)

# To clear the cache and change the value you need to delete the cache
del ec.computation_result
print('*'*20)
print(ec.computation_result)
```

<div id="deques">
    <h2>deques</h2>
    

    
To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>
</div>

***

- if you need to add and reomve lements from a list quickly and a lot deque is quicker than list
- if add or remove to the start of a list it will take O(n) time whereas with deque is takes O(1) time
- more time efficient than a list

```python
from collections import deque
```

```python
q = deque([1,2,3,4,5])
print(q)
```

```python
# add to the deque
q.append(6)

# add element to the start
q.appendleft(0)

# remove element (pop)
x = q.pop() # this will remove from the right
print(q,x) # x is now the popped element 

# pop from the left
x = q.popleft()
print(q,x) # x is now 0
```

```python
import psutil
import time
import functools

def profile_memory(func):
    """
    Decorator to measure memory usage and execution time of a function.
    """
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        # Get initial memory usage
        process = psutil.Process()
        mem_before = process.memory_info().rss / 1e6  # Convert to MB
        start_time = time.time()

        # Run the function
        result = func(*args, **kwargs)

        # Get final memory usage and execution time
        mem_after = process.memory_info().rss / 1e6  # Convert to MB
        end_time = time.time()
        
        print(f"Function '{func.__name__}' stats:")
        print(f"Memory Usage: {mem_after - mem_before:.2f} MB")
        print(f"Execution Time: {end_time - start_time:.2f} seconds")

        return result

    return wrapper
```

```python
@profile_memory
def example_function():
    """Simulate some computation with memory usage."""
    large_list = [x**2 for x in range(10**10)]  # Simulates memory usage
    time.sleep(2)  # Simulates processing time
    return sum(large_list)

# Call the function with profiling
result = example_function()
result
```

## dis
- dissassemble package
- show the C code ran for python code submitted 
- example here showing why list comphrehension is generally better than for loop for appending to a list

```python
import dis
```

```python
def for_loop(n):
    my_list = []
    for x in range(n):
        my_list.append(x)
    return my_list

def  list_comp(n):
    return [x for x in range(n)]

print(dis.dis(for_loop))
print('*'*30)
print(dis.dis(list_comp))
```

- both operations start at FOR_ITER for the iterations in both operations
- STORE_FAST stores the next item in a variable called x
- LOAD_FAST in the for loop it loads the entire my_list whereas in the list comphrehension i only loads the current value x
- The for loop then has to load the LOAD_METHOD function (sometimes called LOAD_ATTR), then has to LOAD_FAST to then CALL_METHOD in order to append the current value 
- The for loop has to realise it is a list, then get the appropriate function for the list and then perform the action so it has additional steps
- The list comphrehesion already knows it is dealing with a list and the appropriate method to append so can call LIST_APPEND (specialised byte code)
- note, if you are performing multiple operations within the list comphrehension then it may be more appropriate to use a for loop for readability - too keep in mind
