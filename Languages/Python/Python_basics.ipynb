#### Assignment expression operator

Operator used to evaluate, assign, and return value from a single statement in Python

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# Walrus -    variable := expression

# basic loop
num=int(input("Enter a number"))
while num > 0:
    print(num)
    num=num-1
    
# Using walrus
number = int(input("Enter a number: "))
while number > 0:
    print(number)
    (number := number - 1)
```

#### Default dictionaries

```python
from collections import defaultdict

# Example 1 - auto initialising keys and assigning default or specialised value
my_dict = defaultdict(int)

my_dict['age'] # as age doesn't exist it will automatically assign age to have the value 0

my_dict['age2'] += 20 # assign age2 as the value 20

print(my_dict)

# Example 2
my_list = [1,2,3,4,5,6,7,8,3,65,34,5,1]
counter = defaultdict(int)

for element in my_list:
    counter[str(element)] += 1

print(counter)

# list
words = ['apple','banana','grapes','kiwi','pear','melon','avocado']

grouped_words = defaultdict(list)

for word in words:
    grouped_words[word[0]].append(word)
    
print(grouped_words)

# tuple

tuple_list = [("A", 10),("B", 4),("C", 1),("D", 7),("A",5),("B",20)]

grouped_data = defaultdict(list)

for key,value in tuple_list:
    grouped_data[key].append(value)
    
grouped_data = {k: sum(v) for k, v in grouped_data.items()}
print(grouped_data)

# own defaultdict

class MyDefaultDict(defaultdict):
    
    def __missing__(self, key):
        self[key] = value = len(key)
        return value
    
test = MyDefaultDict()

print(test["hello"])
print(test["hi"])
print(test)


# constant defaultdict / lambda expression defaultdict

constant_default_dict = defaultdict(lambda: "hello world") # assigned the default value
print(constant_default_dict["hello"])
```

### Classes

- A class is a blueprint for creating objects, and objects are instances of classes
- Classes allow you to define attributes (data) and methods (functions) that operate on those attributes

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# A simple class
class Dog:
    # Class-level attribute
    species = "Canis familiaris"

    # Constructor method (initialization)
    def __init__(self, name, age):
        self.name = name  # Instance-level attribute
        self.age = age

    # Instance method
    def description(self):
        return f"{self.name} is {self.age} years old."

    # Another instance method
    def speak(self, sound):
        return f"{self.name} says {sound}"

# Creating instances (objects) of the Dog class
dog1 = Dog("Buddy", 3)
dog2 = Dog("Milo", 5)

# Accessing attributes and calling methods
print(dog1.name)         # Output: "Buddy"
print(dog2.description())  # Output: "Milo is 5 years old."
print(dog1.speak("Woof"))  # Output: "Buddy says Woof"
```

```python
# You can have a class inherit information from a parent class
class GoldenRetriever(Dog):
    def speak(self, sound="Bark"):
        return f"{self.name} says {sound}"

golden = GoldenRetriever("Max", 2)

print(golden.species)      # Inherited from the parent class
print(golden.speak("Woof"))  # Overridden method in the subclass
```

```python
# Class attributes are shared among all instances of the class, while instance attributes are specific to each instance
class Car:
    # Class attribute
    category = "Automobile"

    def __init__(self, make, model):
        # Instance attributes
        self.make = make
        self.model = model

car1 = Car("Toyota", "Camry")
car2 = Car("Honda", "Civic")

print(car1.category)  # Output: "Automobile" (Class attribute)
print(car2.make)      # Output: "Honda" (Instance attribute)
```

#### Static and class methods

- These are both Decorators
- Static methods are bound to the class and don't depend on instance-specific data
- Class methods are bound to the class and can access and modify class-level attributes
- Static and class methods can be called using the class itself or an instance of the class
- You use the @staticmethod and @classmethod decorators to define static and class methods, respectively

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# Static methods and class methods are methods that can be called on the class itself, rather than on instances of the class
class MathUtils:
    @staticmethod
    def add(x, y):
        return x + y

    @classmethod
    def multiply(cls, x, y):
        return cls.add(x, y) * 2

result1 = MathUtils.add(3, 4)       # Static method call
result2 = MathUtils.multiply(3, 4)  # Class method call

print(result1)  # Output: 7
print(result2)  # Output: 14
```

**Static method further explanation:**

A static method is a method that belongs to a class rather than an instance of the class. It is defined using the @staticmethod decorator and does not take the instance (self) or class (cls) as its first parameter. Static methods are often used for utility functions that do not depend on instance-specific data.

**Class method further explanation:**

A class method is a method that is bound to the class and not the instance. It is defined using the @classmethod decorator and takes the class itself (cls) as its first parameter. Class methods are often used for operations that involve the class but don't necessarily require an instance.

### Decorators

- Decorators are a powerful and flexible way to modify or enhance the behavior of functions or methods without changing their code
- Decorators are typically used to add functionality to functions, such as logging, authentication, or memoization

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# simple decorator
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python
# decorator with arguments
def repeat(num):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(num):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(3)
def say_hello(name):
    print(f"Hello, {name}!")

say_hello("Alice")
```

#### Class based decorator

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
import time

class TimingDecorator:
    def __init__(self, func):
        self.func = func

    def __call__(self, *args, **kwargs):
        start_time = time.time()
        result = self.func(*args, **kwargs)
        end_time = time.time()
        print(f"{self.func.__name__} took {end_time - start_time} seconds to execute.")
        return result

@TimingDecorator
def slow_function():
    time.sleep(2)

slow_function()
```

#### Custom iterators

- Used within clases using __iter__() and __next__()

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
class MyIterator:
    def __init__(self, max_limit):
        self.max_limit = max_limit
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.max_limit:
            self.current += 1
            return self.current
        raise StopIteration

my_iter = MyIterator(3)

for num in my_iter:
    print(num)
```

```python
def raise_various_exceptions(choice):
    try:
        if choice == 1:
            # Raise a ValueError
            raise ValueError("This is a custom ValueError.")
        elif choice == 2:
            # Raise a generic Exception
            raise Exception("This is a generic Exception.")
        elif choice == 3:
            # Raise a ZeroDivisionError
            result = 10 / 0
        else:
            # Raise an IndexError
            my_list = [1, 2, 3]
            value = my_list[4]
    except ValueError as ve:
        print(f"Caught a ValueError: {ve}")
    except Exception as e:
        print(f"Caught a generic Exception: {e}")
    except ZeroDivisionError as zde:
        print(f"Caught a ZeroDivisionError: {zde}")
    except IndexError as ie:
        print(f"Caught an IndexError: {ie}")
    except:
        print("Caught an unexpected error.")
    else:
        print("No exceptions were raised.")
    finally:
        print("This code always runs, regardless of exceptions.")

# Test different cases to raise various exceptions
print("--- Case 1: ValueError ---")
raise_various_exceptions(1)

print("\n--- Case 2: Generic Exception ---")
raise_various_exceptions(2)

print("\n--- Case 3: ZeroDivisionError ---")
raise_various_exceptions(3)

print("\n--- Case 4: IndexError ---")
raise_various_exceptions(4)
```

### Yield over return

- Return a value and then keep the function going can be achieved using yield
- The function which uses yield is known as a generator function

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
def hello():
    yield 1
    yield 2
    yield 3
    
for i in hello():
    print(i)
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a  # Yield current Fibonacci number
        a, b = b, a + b  # Update values

# Using the generator
for num in fibonacci(10):  # Generate first 10 Fibonacci numbers
    print(num)
```
