```python
from icecream import ic
```

```python
def multiply(a, b):
    result = a * b
    ic(result)
    return result
multiply(2, 3)

# Output: ic| result: 6
```

```python
def multiply_and_divide(a, b):
    result = a * b
    ic(result, a/b)
    return result, a/b
multiply_and_divide(2, 3)

# Output: ic| result: 6, a/b: 0.6666666666666666
```

```python
def is_even(num):
    result = num % 2 == 0
    ic(result, num)
    return result
is_even(4)
is_even(5)

# Outputs
ic| result: True, num: 4
ic| result: False
```

```python
def sum_list(numbers):
    result = sum(numbers)
    ic(numbers, result)
    return result
sum_list([1, 2, 3, 4, 5])

# Output
ic| numbers: [1, 2, 3, 4, 5], result: 15
```
