```python
# Basic
wget [URL]

# Download file with progress bar only
wget -q --show-progress [URL] 

# Download file and save with a different name
wget -O [FILENAME] [URL]

# Download a file to a specific directory
wget -P [DIRECTORY] [URL]

# Resume a partially downloaded file
wget -c [URL]

# Retry a failed download multiple times
wget --tries=10 [URL]

# Download multiple files by listing URLs in a text file - (Each line in urls.txt should contain one URL.)
wget -i urls.txt

# Download files using a URL pattern (e.g., sequentially numbered files)
wget http://example.com/file-{1..10}.jpg

# Download a single webpage with all necessary resources (images, CSS, JS, etc.)
wget -p --convert-links [URL]

# Download an entire website for offline viewing
wget -m -k -E -p --no-parent [URL]

# Mirror a website but limit recursion depth
wget -r -l2 [URL]

# Download a file quietly (no output)
wget -q [URL]

# Limit download speed (e.g., 500 KB/s)
wget --limit-rate=500k [URL]

# Set a custom User-Agent string
wget -U "Mozilla/5.0" [URL]

# Download only response headers (without saving the file)
wget --spider -S [URL]

# Authenticate with a username and password
wget --user=[USERNAME] --password=[PASSWORD] [URL]

# Set a custom referrer URL
wget --referer=[REFERRER_URL] [URL]

# Download all files in a directory if indexing is enabled - (This downloads all .jpg, .png, and .gif files from the directory.)
wget -r -l1 -A .jpg,.png,.gif [URL]

# from a sever using a token
wget https://server.ac.uk/user/{username}/files/path/to/file/?token={token}

# to retrieve a file from google drive
- go to the file of interest and check it has public/open access link
- copy the download link / share link and paste in a text file
- extract the id (always starts with 1) from the download link and insert in below structure replacing {id}

wget --load-cookies /tmp/cookies.txt "https://docs.google.com/uc?export=download&confirm=$(wget --quiet --save-cookies /tmp/cookies.txt --keep-session-cookies --no-check-certificate 'https://docs.google.com/uc?export=download&id={id}' -O- | sed -rn 's/.*confirm=([0-9A-Za-z_]+).*/\1\n/p')&id={id}" -O file_name.extension && rm -rf /tmp/cookies.txt

# to retrieve a file from one drive (buisness)
- go to the file of interest and check it has public/open access link
- copy the share link and paste in a text file
- add ?download=1 to the end of the link and add insert in below structure replacing {id}

wget --no-check-certificate --content-disposition -O "file_name.extension" "{id}"
```
