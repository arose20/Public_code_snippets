### Basic environment commands

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# list environments
conda env list
conda info --envs

# activate enviornment
conda activate env
source activate env

# remove an conda enviornment
conda remove --name myenv --all

# list packages within virtual enviornment which you are located in
conda list

# to see if specific package is installed in conda enviornment 
conda list -n myenv package

# to update conda packages
conda env update
```

### Sharing environments

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# sharing enviornments 

# exporting environmnet you are currently in (exports all packages within enviornment)
conda env export > environment.yml

# to export an enviornment which contains only the specific package you installed when creating enviornment e.g. conda create -n myenv python=3.6 scipy=0.15.0
conda env export --from-history # in this case it exports an yml file with python3.6 and scipy0.15.0

# create new enviornment using yml file
conda env create -f environment.yml


# to clone an enviornemnt
conda create --name myclone --clone myenv

# to build an identical environment

# first build a spec list into a text file
conda list --explicit > spec-file.txt

#To use the spec file to create an identical environment on the same machine or another machine:

conda create --name myenv --file spec-file.txt

#To use the spec file to install its listed packages into an existing environment:

conda install --name myenv --file spec-file.txt

# add prune command to above(s)
```
