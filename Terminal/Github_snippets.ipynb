### Basic setup

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# set the author name and email address respectively to be used with your git commits
git config --global user.name "github_username"
git config --global user.email "desired_email@gmail.com"

# set the command line coloring for Git for easy reviewing
git config --global color.ui auto

# system wide ignore pattern for all local repositories
git config --global core.excludesfile [file]

# get global config 
git config --global --list
```

### Stage and snapshot

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# To list all files that have to be committed and if there is any changes from the repo - won't tell you exact changes
git status

# to add a single file your will want to add to your repo stored in a "staging area"
git add [file]

# to add one or more files to the staging area
git add *

# to unstage a file while preserving its content
git reset [file]

# to undo all commits after a specified commit and preserve changes locally
git reset commit     # e.g. git reset 09bb8e3f996eaf9a68ac5ba8d8b8fceb0e8641e7

# To discard all history and go back to a specified commit
git reset --hard commit

# To apply any commits of the current branch ahead of a specified one
git rebase [branch]

# to show which files are not staged yet - shows you all the changes 
git diff

# to show differences in files in the staging area and the latest version present
git diff --staged

# to show the differences between two branches
git diff [branch_1] [branch_2]

# to record or snapshot a file permentantly in the version history (actually adding to the repo)
git commit -m "message to describe what is being committed"

# commit any files you have added with git add and also commit any files changed since then 
git commit -a

# To revert commit changes
git revert commitid
```

### Branch and merge

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# To list all branches in the current repositiry
git branch

# To list all branches local and remote
git branch -a

# To create a new branch 
git branch [branch_new]

# To delete a branch
git branch -d [branch_new]

# To delete a branch forcefully
git branch -D [branch_new]

# To delete a remote branch
git push orign --delete [branch]

# To switch between branches
git checkout -b [branch_V5]

# To clone a remote branch and switch to it
git checkout -b [branch] origin/[branch]

# To switch to the branch last checked out
git checkout -

# To discard changes to a file
git checkout -- file

# To merge a specified branchs history with another branch i.e. another branch into the branch you are currently in
git merge [branch_2] # this merges branch_2 into the branch you are in

# To rename a local branch
git branch -m [old-branch-name] [new branch-name]

# To tag a specific commit 
git tag commit
```

### Insepct and compare

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# To list the version history for the current branch
git log

# To view changes detailed
git log --summary

# To view changes briefly
git log --oneline

# To list the version history for a file, including the renaming of files aswell
git log --follow [file]

# To show the commits on branch_1 that are not on branch_2
git log [branch_2] [branch_1]

# To show all commit logs with indication of any paths that moved
git log --stat -M

# To show the metadata and content changes for a specified commit
git show commit
```

### Share and update

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# To fetch changes on the remote server to your local working directory
git fetch [Repo]

# To fetch and merge any changes on the remote server to your local working directory
git pull [Repo]

# To pull changes from remote repository
git pull origin [branch]

# To connect your local repository to the remote server 
git remote add [variable] [server_link]  # e.g. git remote add origin https://github.com/arose20/GitDemo.git

# To send the committed changes of the branch you are in to the repository
git push

# To push a branch to your remote repository 
git push origin [branch]

# To push changes to remote repository and remember the branch
git push -u origin [branch]

# To send the committed changes of the master branch to the repository
git push [variable] master 

# To send the committed changes of the branch you are in to the repository
git push [variable] [branch_2]

# To push all branches to your remote repository
git push -all [variable]

# To push your changes while ignoring any potential conflicts and make your version the final version
git push --force [variable] [branch]
    
# To push while ignoring verification
git push --no-verify
```

### Tracking path changes

### Temporary commits

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# To temporarily store all the modified tracked files
git stash
git stash save

# To restore the most recently stashed files
git stash pop

# To list all stashed changesets
git stash list

# To discard the most recently stashed changeset
git stash drop

# To remove all stashed entries
git stash clear
```

### Setup with ssh

To go back to the main menu click <a href="#document-information" style="color: blue;">here</a>

```python
# 1. ssh to server
ssh username@server.ac.uk

# 2. Set up a ssh key (github linux tutorial steps) : https://docs.github.com/en/authentication/connecting-to-github-with-ssh/generating-a-new-ssh-key-and-adding-it-to-the-ssh-agent
ssh-keygen -t ed25519 -C "your_email@example.com"

# 3. Use default path so it is in home directory for private directory or supply desired directory

# 4. Optional: enter a passphrase for security, hit enter to leave blank if don't want password

# 5. Add ssh key to shh agent
eval "$(ssh-agent -s)"
ssh-add ~/.ssh/id_ed25519


# 6. Add ssh key to github account : https://docs.github.com/en/authentication/connecting-to-github-with-ssh/adding-a-new-ssh-key-to-your-github-account

# Command line method
gh ssh-key add ~/.ssh/id_ed25519.pub

# Web browser method
# print the key and copy the ouput
cat ~/.ssh/id_ed25519.pub

# 7. Add key and authenticate on github account

# go to github profile > Settings > ssh keys > new ssh key > title (server_name) , paste key > add ssh key > authenticate


# 8. Configure github account username and email on server
git config --global user.email "desired_email@gmail.com"
git config --global user.name "github_username"
```

```python
# Add to local if mount server

# 1. Configure github account username and email locally
git config --global user.email XXX

# 2. Make a ssh folder if don't have one
mkdir .ssh

# 3. Using a code editor to open the config file - below uses vim
vim .ssh/config

# 4. Add Github information to config file
Host github.com
    User git
    IdentityFile /server/users/user/.ssh/id_ed25519   # replace with id key and path to your user profile

# 5. Exit and save changes
:x
    
# 6. Make sure username and email is setup
git config --global user.email "desired_email@gmail.com"
git config --global user.name "github_username"
```
