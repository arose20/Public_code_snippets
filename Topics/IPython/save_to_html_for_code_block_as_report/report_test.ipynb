```python
run_and_capture_all("example_3", """
import numpy as np
import matplotlib.pyplot as plt

# ---- Plot 1: Line plot ----
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='sin(x)')
plt.title("Sine Wave")
plt.legend()
plt.show()  # displayed in notebook and captured separately

# ---- Plot 2: Box plot ----
data = np.random.normal(0, 1, 100)
plt.boxplot(data)
plt.title("Random Boxplot")
plt.show()  # independent figure

# ---- Plot 3: Violin plot ----
data2 = [np.random.normal(mu, 0.5, 100) for mu in [0, 1, 2]]
plt.violinplot(data2)
plt.title("Random Violin Plot")
plt.show()  # independent figure

print("All 3 plots generated successfully")
""", display_live=True)
```
