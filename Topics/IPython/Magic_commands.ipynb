# Store magic example
- create a variable in one session and pass it into another session

```python
# Create a variable
my_variable = "Hello, passing to another session!"

# Use %store to persistently store the variable
%store my_variable
```

```python
# List all stored variables
%store
```

```python
# Clear all stored variables
%store -z
```

```python
# List all stored variables
%store
```

```python
import pandas as pd

# Create a variable
my_variable_1 = "Hello, passing to another session!"
my_variable_2 = "Hello, passing to another session 2!"
my_variable_3 = pd.DataFrame()

# Use %store to persistently store the variable
%store my_variable_1
%store my_variable_2
%store my_variable_3
```

```python
# List all stored variables
%store
```

```python
# Capture the output of %store
%store -r

# Assuming %who_ls has been executed in an IPython or Jupyter environment
interactive_variables = %who_ls

# Convert to a list
variable_list = list(interactive_variables)

# Now, variable_list contains the names of all stored variables
print(variable_list)
```

```python
# Clear all stored variables
%store -z
```
