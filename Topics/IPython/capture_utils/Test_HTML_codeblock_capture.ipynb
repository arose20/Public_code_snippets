```python
from capture_utils import run_and_capture_all, show_capture_all, save_capture_to_html
```

```python
# Cell 2: Example 1 - Simple print statements (live display)
code_1 = """
print("Hello world!")
for i in range(3):
    print(f"Counting: {i+1}")
"""

# Run and capture
run_and_capture_all("example_1", code_1, display_live=True)

# Save to HTML
save_capture_to_html("example_1", output_file="example_output.html", title="Example 1", include_timestamp=True)
```

```python
# Cell 3: Example 2 - Silent capture with later inspection
code_2 = """
import math
results = [math.sqrt(x) for x in range(10)]
print("Square roots computed!")
results
"""

# Run silently (no live output)
run_and_capture_all("example_2", code_2, display_live=False)

# Inspect captured output
show_capture_all("example_2", section_title="Example 2 Output")

# Save to HTML
save_capture_to_html("example_2", output_file="example_output.html", title="Example 2", include_timestamp=False)
```

```python
# Cell 4: Example 3 - Rich output (matplotlib plot)
code_3 = """
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure()
plt.plot(x, y, label='sin(x)')
plt.title("Sine Wave")
plt.legend()
plt.show()

print("Plot generated successfully")
"""

# Run and capture
run_and_capture_all("example_3", code_3, display_live=True)

# Save to HTML
save_capture_to_html("example_3", output_file="example_output2.html", title="Example 3 - Plot")
```

```python
# Cell 5: Show all captures in notebook
show_capture_all("example_1", section_title="Example 1 Output")
show_capture_all("example_2", section_title="Example 2 Output")
show_capture_all("example_3", section_title="Example 3 Output")
```
